# **Copy and download tool*****

```python
# copy or delete in output
import os
import shutil
import requests

def main():
    print("What action would you like to perform?")
    print("1. Copy a file")
    print("2. Copy a folder")
    print("3. Delete a file/folder")
    print("4. Download from URL")
    print("5. Save text from input")
    print("6. view text in output")
    action = input("Enter the number of your choice (1/2/3/4/5/6): ").strip()

    if action == "1":
        src = input("Enter the full path to the source file: ").strip()
        dest = input("Enter the full path to the destination file: ").strip()
        try:
            shutil.copy2(src, dest)
            print(f"✅ File copied from:\n{src}\nto:\n{dest}")
        except Exception as e:
            print(f"❌ Error copying file: {e}")

    elif action == "2":
        src = input("Enter the full path to the source folder: ").strip()
        dest = input("Enter the full path to the destination folder: ").strip()
        try:
            shutil.copytree(src, dest)
            print(f"✅ Folder copied from:\n{src}\nto:\n{dest}")
        except Exception as e:
            print(f"❌ Error copying folder: {e}")

    elif action == "3":
        target = input("Enter the full path to the file or folder to delete: ").strip()
        if os.path.isfile(target):
            try:
                os.remove(target)
                print(f"✅ File deleted: {target}")
            except Exception as e:
                print(f"❌ Error deleting file: {e}")
        elif os.path.isdir(target):
            try:
                shutil.rmtree(target)
                print(f"✅ Folder deleted: {target}")
            except Exception as e:
                print(f"❌ Error deleting folder: {e}")
        else:
            print("⚠️ Path does not exist or is not accessible.")
            
    elif action == "4":
        src = input("Enter the full URL to the source file: ").strip()
        dest = input("Enter the full path to the destination file name: ").strip()
        try:
            print(f"Downloading from {src}...")
            response = requests.get(src, stream=True)
            response.raise_for_status()
    
            with open(dest, "wb") as f:
                for chunk in response.iter_content(chunk_size=8192):
                    f.write(chunk)
    
            print(f"✅ Download completed and saved as '{dest}'")
            # # Path to your audio file
            # # url = "https://www.divinerevelations.info/documents/bible/laotian_mp3_bible_nt/lao_ssr_nt_non_drama/_chapters_/01_Matthew/Matthew_02.mp3"
            # # local_path = "/kaggle/working/audio02.mp3"
            # # Download the file
            # with open(src, "wb") as f:
            #     f.write(dest.get(url).content)
            # print(f"✅ Downloaded to \nto:\n{dest}")
            
        except Exception as e:
            print(f"❌ Error downloading the file: {e}")
            
    elif action == "5":
        print("Enter/paste your text. Press ENTER twice to finish:")
        lines = []
        while True:
            line = input()
            if line == "":
                break
            lines.append(line)
    
        text = "\n".join(lines)
        dest = input("Enter the destination filename (e.g. transcript.txt): ").strip()
    
        try:
            with open(dest, "w", encoding="utf-8") as f:
                f.write(text)
            print(f"✅ Text successfully saved to '{dest}'")
        except Exception as e:
            print(f"❌ Error saving file: {e}")
            
    elif action == "6":
        filename = input("Enter the full path to the source file: ").strip()
        
        try:
            
            if not os.path.exists(filename):
                print("❌ File not found.")
            elif filename.endswith(".json"):
                with open(filename, "r", encoding="utf-8") as f:
                    data = json.load(f)
                print("✅ JSON loaded:")
                print(data)
            elif filename.endswith(".txt"):
                with open(filename, "r", encoding="utf-8") as f:
                    data = f.read()
                print("✅ Text file content:")
                print(data)
            else:
                print("⚠️ Unsupported file type.")
        except Exception as e:
            print(f"❌ Error cannpt read file: {e}")
            
    else:
        print("⚠️ Invalid choice. Please enter 1, 2, 3, 4, 5 or 6.")
    

# Run the function
main()
```

```python
from pydub import AudioSegment
import os

input_path = "output_audio/vocals.wav"
output_path = "mms_ready/vocals_16k.wav"

# Make sure output directory exists
os.makedirs(os.path.dirname(output_path), exist_ok=True)

# Load and convert
audio = AudioSegment.from_wav(input_path)
audio = audio.set_channels(1)       # mono
audio = audio.set_frame_rate(16000) # 16kHz
audio = audio.set_sample_width(2)   # 16-bit PCM

# Export
audio.export(output_path, format="wav")
```
